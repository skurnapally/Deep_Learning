```python
from transformers import pipeline
```

```python
pipe = pipeline(task = 'text2text-generation', model = 'facebook/m2m100_418M')
```

```python
pipe('hi how are you ?',forced_bos_token_id = pipe.tokenizer.get_lang_id(lang = 'ta'))
```
